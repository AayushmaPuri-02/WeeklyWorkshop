```python
def add_binary(a, b):
    """
    Returns the sum of two decimal numbers in binary digits.

    Parameters:
    a (int): A decimal integer
    b (int): Another decimal integer

    Returns:
    binary_sum (str): Binary string of the sum of a and b
    """
    # Sum the integers and convert to binary string using bin()
    # [2:] removes the '0b' prefix that Python adds to binary strings
    binary_sum = bin(a + b)[2:]
    return binary_sum

result = add_binary(5, 3)
print(result)  # Output: '1000'

```

**User - Defined - Function**

This function converts temperature between Celsius and Fahrenheit.
This function prompts the user to specify the conversion type (Celsius to Fahrenheit or
Fahrenheit to Celsius)
and then asks for the temperature value. It calculates and returns the converted temperature.
Returns:
float: The converted temperature value.

```python
def temperature_converter():

  print("Choose conversion type:")
  print("1. Celsius to Fahrenheit")
  print("2. Fahrenheit to Celsius")
  # Get conversion choice from user
  choice = input("Enter 1 or 2: ")
  if choice == "1":
  # Celsius to Fahrenheit conversion
    celsius = float(input("Enter temperature in Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}C is equal to {fahrenheit}F")
    return fahrenheit
  elif choice == "2":
    # Fahrenheit to Celsius conversion
    fahrenheit = float(input("Enter temperature in Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"{fahrenheit}F is equal to {celsius}C")
    return celsius
  else:
    print("Invalid choice. Please enter 1 or 2.")
    return None
# Call the function
temperature_converter()
```

**Task - 2:**
Create a Python program that performs various mathematical operations on a list of numbers.
• The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).
3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The maximum number in the list.
    """
    return max(numbers)


def find_min(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The minimum number in the list.
    """
    return min(numbers)


def main():
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1-4): ")

    try:
        numbers_input = input("Enter numbers separated by spaces: ")
        numbers = [float(num) for num in numbers_input.split()]

        if len(numbers) == 0:
            raise ValueError("Empty list")

        if choice == "1":
            result = calculate_sum(numbers)
            print("Sum:", result)

        elif choice == "2":
            result = calculate_average(numbers)
            print("Average:", result)

        elif choice == "3":
            result = find_max(numbers)
            print("Maximum:", result)

        elif choice == "4":
            result = find_min(numbers)
            print("Minimum:", result)

        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Invalid input. Please enter numeric values separated by spaces.")


main()
```

**1. Flatten a Nested List:**
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
• Requirements:
– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.
– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): A nested list.

    Returns:
        list: A flattened list containing all elements.
    """
    result = []

    for sublist in lst:
        for item in sublist:
            result.append(item)

    return result

numbers = [[1, 2], [3, 4], [5]]
print(flatten(numbers))
```

**4. Remove Specific Element from a Nested List:**
Write a Python function that removes all occurrences of a specific element from a nested list.
• Requirements:
– Define a function remove element(lst, elem) that removes elem from lst and returns the
modified list.
– Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should
be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from a nested list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result

lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

print(remove_element(lst, elem))
```

**5. Find the Maximum Element in a Nested List:**
Write a Python function that finds the maximum element in a nested list (regardless of depth).
• Requirements:
– Define a function find max(lst) that takes a nested list lst and returns the maximum
element.
– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Returns the maximum number in a nested list.
    """
    maximum = float('-inf')

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)
            if value > maximum:
                maximum = value
        else:
            if item > maximum:
                maximum = item

    return maximum

data = [[1, 2], [3, [4, 5]], 6]
print(find_max(data))
```

**7. Flatten a List of Lists of Lists:**
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.
• Requirements:
– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
flattened list.
– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into one single list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result

data = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(data))
```

**8. Nested List Average:**
Write a Python function that calculates the average of all elements in a nested list.
• Requirements:
– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.
– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """
    Returns the average of all numbers in a nested list.
    """
    total = 0
    count = 0

    for item in lst:
        if isinstance(item, list):
            sub_total, sub_count = average_nested(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1

    return total, count


data = [[1, 2], [3, 4], [5, 6]]

total, count = average_nested(data)
average = total / count

print(average)
```

```python
import numpy as np
# Creating two 3D arrays
arr1 = np.array([[[1], [2]], [[3], [4]]]) # Shape: (2, 2, 1)
arr2 = np.array([[[5], [6]], [[7], [8]]]) # Shape: (2, 2, 1)
# Concatenate along axis 0 (vertical stacking)
concat_axis0 = np.concatenate((arr1, arr2), axis=0)
# Shape of concat_axis0: (4, 2, 1), arr1 is stacked on top of arr2
# Concatenate along axis 1 (horizontal stacking)
concat_axis1 = np.concatenate((arr1, arr2), axis=1)
# Shape of concat_axis1: (2, 4, 1), arr1 and arr2 are placed side by side
# Concatenate along axis 2 (depth stacking)
concat_axis2 = np.concatenate((arr1, arr2), axis=2)
# Shape of concat_axis2: (2, 2, 2), arr1 and arr2 are stacked in depth (third
#dimension)
# Displaying the results
print("Concatenation along axis 0 (Vertical stacking):\n", concat_axis0)
print("Shape of concatenated array along axis 0:", concat_axis0.shape)
print("\nConcatenation along axis 1 (Horizontal stacking):\n", concat_axis1)
print("Shape of concatenated array along axis 1:", concat_axis1.shape)
print("\nConcatenation along axis 2 (Depth stacking):\n", concat_axis2)
print("Shape of concatenated array along axis 2:", concat_axis2.shape)
```

1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}

```python
# 10.1 Basic Vector and Matrix Operations with NumPy
# Import NumPy library (required before using NumPy functions)
import numpy as np

# 1. Initialize an empty array with size 2x2
empty_array = np.empty((2, 2))
# (Creates a 2x2 array without initializing entries. Values are random garbage from memory.)

print("1. Empty 2x2 array:\n", empty_array)


# 2. Initialize an all-one array with size 4x2
ones_array = np.ones((4, 2))
# (Creates a 4x2 matrix where every element is 1)

print("\n2. 4x2 array filled with ones:\n", ones_array)


# 3. Return a new array of given shape and type, filled with a specific value
full_array = np.full((3, 3), 7)
# (Creates a 3x3 matrix where every element is the value 7)

print("\n3. Array filled with a specific value (7):\n", full_array)


# 4. Return a new array of zeros with the same shape and type as a given array
zeros_like_array = np.zeros_like(full_array)
# (Creates a new array with same shape as full_array but filled with 0)

print("\n4. Zeros array with same shape as full_array:\n", zeros_like_array)


# 5. Return a new array of ones with the same shape and type as a given array
ones_like_array = np.ones_like(full_array)
# (Creates a new array with same shape as full_array but filled with 1)

print("\n5. Ones array with same shape as full_array:\n", ones_like_array)


# 6. Convert an existing list into a NumPy array
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
# (Converts the Python list into a NumPy array)

print("\n6. Converted list to NumPy array:\n", numpy_array)
```

**Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:**
Complete the following tasks:
1. Create an array with values ranging from 10 to 49. {Hint:np.arrange()}.
2. Create a 3X3 matrix with values ranging from 0 to 8.
{Hint:look for np.reshape()}
3. Create a 3X3 identity matrix.{Hint:np.eye()}
4. Create a random array of size 30 and find the mean of the array.
{Hint:check for np.random.random() and array.mean() function}
5. Create a 10X10 array with random values and find the minimum and maximum values.
6. Create a zero array of size 10 and replace 5th element with 1.
7. Reverse an array arr = [1,2,0,0,4,0].
8. Create a 2d array with 1 on border and 0 inside.
9. Create a 8X8 matrix and fill it with a checkerboard pattern.

```python
# Problem 2: Array Manipulation – Numerical Ranges and Array Indexing

import numpy as np


# 1. Create an array with values ranging from 10 to 49
arr1 = np.arange(10, 50)
# (np.arange(start, end) generates numbers starting from 10 up to 49)

print("1. Array from 10 to 49:\n", arr1)


# 2. Create a 3x3 matrix with values ranging from 0 to 8
arr2 = np.arange(9).reshape(3, 3)
# (First create numbers 0–8, then reshape into a 3x3 matrix)

print("\n2. 3x3 matrix from 0 to 8:\n", arr2)


# 3. Create a 3x3 identity matrix
arr3 = np.eye(3)
# (Identity matrix = diagonal values are 1 and everything else is 0)

print("\n3. 3x3 Identity Matrix:\n", arr3)


# 4. Create a random array of size 30 and find the mean
arr4 = np.random.random(30)
# (Creates 30 random numbers between 0 and 1)

mean_value = arr4.mean()
# (Calculates the average of the array)

print("\n4. Random array of size 30:\n", arr4)
print("Mean of the array:", mean_value)


# 5. Create a 10x10 array with random values and find min and max
arr5 = np.random.random((10, 10))
# (Creates a 10x10 matrix filled with random numbers)

min_val = arr5.min()
max_val = arr5.max()

print("\n5. 10x10 random array:\n", arr5)
print("Minimum value:", min_val)
print("Maximum value:", max_val)


# 6. Create a zero array of size 10 and replace 5th element with 1
arr6 = np.zeros(10)
# (Creates an array with 10 zeros)

arr6[4] = 1
# (Index starts from 0, so the 5th element is index 4)

print("\n6. Zero array with 5th element replaced by 1:\n", arr6)


# 7. Reverse an array
arr7 = np.array([1, 2, 0, 0, 4, 0])

reversed_arr = arr7[::-1]
# ([::-1] reverses the order of elements in the array)

print("\n7. Reversed array:\n", reversed_arr)


# 8. Create a 2D array with 1 on border and 0 inside
arr8 = np.ones((5, 5))
# (Start with a 5x5 array filled with 1)

arr8[1:-1, 1:-1] = 0
# (Replace the inner area with 0, keeping borders as 1)

print("\n8. 2D array with border 1 and inside 0:\n", arr8)


# 9. Create an 8x8 checkerboard matrix
arr9 = np.zeros((8, 8), dtype=int)
# (Start with an 8x8 matrix filled with 0)

arr9[1::2, ::2] = 1
arr9[::2, 1::2] = 1
# (Fill alternate positions with 1 to create checkerboard pattern)

print("\n9. 8x8 Checkerboard pattern:\n", arr9)
```

For the following arrays:

Complete all the task using numpy:

2. Subtract the two array.

4. Find the square of each element of the array.
5. Find the dot product between: v(and)w ; x(and)v ; x(and)y.

{Hint:try np.concatenate() or np.vstack() functions.

```python
# Problem 3: Array Operations

import numpy as np

# Given arrays
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11,12])


# 1. Add the two arrays
add_result = x + y
# (Element-wise addition: corresponding elements of x and y are added)

print("1. Addition of x and y:\n", add_result)


# 2. Subtract the two arrays
subtract_result = x - y
# (Element-wise subtraction: elements of y are subtracted from x)

print("\n2. Subtraction of x and y:\n", subtract_result)


# 3. Multiply the array with any integer
multiply_result = x * 3
# (Each element of x is multiplied by the integer 3)

print("\n3. Multiplying x with integer 3:\n", multiply_result)


# 4. Square of each element of the array
square_result = np.square(x)
# (Squares every element of x)

print("\n4. Square of each element in x:\n", square_result)


# 5. Dot products
dot_vw = np.dot(v, w)
# (Vector dot product: 9*11 + 10*12)

dot_xv = np.dot(x, v)
# (Matrix-vector multiplication)

dot_xy = np.dot(x, y)
# (Matrix multiplication between x and y)

print("\n5. Dot product v · w:", dot_vw)
print("Dot product x · v:\n", dot_xv)
print("Dot product x · y:\n", dot_xy)


# 6. Concatenate x and y along row
concat_xy = np.concatenate((x, y), axis=0)
# (axis=0 means stacking rows vertically)

print("\n6. Concatenate x and y along rows:\n", concat_xy)

# Concatenate v and w along column
concat_vw = np.vstack((v, w)).T
# (First stack vertically then transpose to make them columns)

print("\nConcatenate v and w along columns:\n", concat_vw)


# 7. Concatenate x and v
try_concat = np.concatenate((x, v))
# (This will raise an error because x is 2D and v is 1D.
# NumPy cannot concatenate arrays with different dimensions.)

print("\n7. Attempt to concatenate x and v:\n", try_concat)
```

```python
# 10.2 Experiment: How Fast is NumPy?
# Compare performance of Python Lists vs NumPy Arrays

import numpy as np
import time
import random


# ------------------------------------------------------------
# 1. Element-wise Addition
# ------------------------------------------------------------

size = 1_000_000

# Python Lists
list1 = [random.random() for _ in range(size)]
list2 = [random.random() for _ in range(size)]

start = time.time()

result_list_add = [list1[i] + list2[i] for i in range(size)]
# (Manually adding elements one-by-one using list comprehension)

end = time.time()

print("1. Python List Addition Time:", end - start, "seconds")


# NumPy Arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()

result_numpy_add = arr1 + arr2
# (NumPy performs vectorized addition internally in optimized C)

end = time.time()

print("1. NumPy Array Addition Time:", end - start, "seconds")


# ------------------------------------------------------------
# 2. Element-wise Multiplication
# ------------------------------------------------------------

# Python Lists
start = time.time()

result_list_mul = [list1[i] * list2[i] for i in range(size)]
# (Multiplying elements one-by-one)

end = time.time()

print("\n2. Python List Multiplication Time:", end - start, "seconds")


# NumPy Arrays
start = time.time()

result_numpy_mul = arr1 * arr2
# (Vectorized multiplication)

end = time.time()

print("2. NumPy Array Multiplication Time:", end - start, "seconds")


# ------------------------------------------------------------
# 3. Dot Product
# ------------------------------------------------------------

# Python Lists
start = time.time()

dot_list = sum(list1[i] * list2[i] for i in range(size))
# (Multiply each pair then sum all values)

end = time.time()

print("\n3. Python List Dot Product Time:", end - start, "seconds")


# NumPy Arrays
start = time.time()

dot_numpy = np.dot(arr1, arr2)
# (Highly optimized C implementation)

end = time.time()

print("3. NumPy Dot Product Time:", end - start, "seconds")


# ------------------------------------------------------------
# 4. Matrix Multiplication (1000 x 1000)
# ------------------------------------------------------------

matrix_size = 1000

# Python Lists
matrix1 = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]
matrix2 = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]

start = time.time()

result_matrix = [[0]*matrix_size for _ in range(matrix_size)]

for i in range(matrix_size):
    for j in range(matrix_size):
        for k in range(matrix_size):
            result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]
# (Classic triple-loop matrix multiplication)

end = time.time()

print("\n4. Python List Matrix Multiplication Time:", end - start, "seconds")


# NumPy Arrays
np_matrix1 = np.array(matrix1)
np_matrix2 = np.array(matrix2)

start = time.time()

result_numpy_matrix = np.dot(np_matrix1, np_matrix2)
# (Uses highly optimized BLAS/LAPACK operations)

end = time.time()

print("4. NumPy Matrix Multiplication Time:", end - start, "seconds")
```
